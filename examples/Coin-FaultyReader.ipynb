```python
# Configure Jupyter so figures appear in the notebook
%matplotlib inline

# Configure Jupyter to display the assigned value after an assignment
%config InteractiveShell.ast_node_interactivity='last_expr_or_assign'

# import classes from thinkbayes2
from thinkbayes2 import Pmf, Suite

import thinkplot
```

```python
class badcoin(Suite):
    
    def Likelihood(self, data, hypo):
        """Computes the likelihood of the data under the hypothesis.
        
        data: heads or tails
        hypo: percentage of heads
        """
        x = hypo/100
        if data== 'H':
            return x*self.y + (1-x)*(1-self.y)
        else:
            return (1-x)*self.y + x*(1-self.y)
```

```python
coin = badcoin(range(101))
thinkplot.Pdf(coin)
thinkplot.Config(xlabel='x', ylabel='PMF')
```

```python
bad_coin = coin.Copy()
good_coin = coin.Copy()
good_coin.y = 1
bad_coin.y = 0.75
for data in 'HHHHHTTT':
    bad_coin.Update(data)
    good_coin.Update(data)
thinkplot.Pdf(bad_coin, color='gray')
thinkplot.Pdf(good_coin)
```
